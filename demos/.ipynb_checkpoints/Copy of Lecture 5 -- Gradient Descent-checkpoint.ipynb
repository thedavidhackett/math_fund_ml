# Lecture 5 -- Gradient Descent

```python
import numpy as np
import numpy.matlib as mat
import numpy.linalg as la
import matplotlib.pyplot as plt

```

### Generate training data

```python
n = 50
p = 1
X = np.matrix(np.random.rand(n, p))
w_true = np.matrix(0.7)
y = X@w_true


```

## Squared-error loss

```python
N = 200
w = np.reshape(np.matrix(np.linspace(0,1,N)),(p,N))
f = np.square(la.norm(mat.repmat(y,1,N) - X@w,2,0))

plt . plot (w.T, f, linewidth =2)
plt . title ('squared error loss for different candidate w')
plt . xlabel ('w')
plt . ylabel ('f(w)')
plt . ylim ([-.5,8.5])
plt . show ()
```

## First gradient 

```python
w_1 = np.matrix(0.2)
f_1 = la.norm(y-X@w_1)**2

gradf_1 = 2*X.T@(X@w_1-y)

tangent_1 = f_1 + gradf_1@(w-w_1)

plt . plot (w.T, f, w_1,f_1,'*',w.T,tangent_1.T,linewidth =2)
plt . title ('squared error loss for different candidate w')
plt . xlabel ('w')
plt . ylabel ('f(w)')
plt . ylim ([-.5,8.5])
plt . show ()
```

## first step

```python
tau = .02
w_2 = w_1 - tau*gradf_1

f_2 = la.norm(y-X@w_2)**2

gradf_2 = 2*X.T@(X@w_2-y)

tangent_2 = f_2 + gradf_2@(w-w_2)

plt . plot (w.T, f, w_2,f_2,'*',w.T,tangent_2.T,linewidth =2)
plt . title ('squared error loss for different candidate w')
plt . xlabel ('w')
plt . ylabel ('f(w)')
plt . ylim ([-.5,8.5])
plt . show ()
```

## multiple iterations


```python
tau = .02
max_iter = 5
w_hat = np.matrix(np.zeros((max_iter+1,1)))
f_hat = np.matrix(np.zeros((max_iter+1,1)))
w_hat[0] = 0.05
f_hat[0] = la.norm(y-X@w_hat[0])**2

for k in range(max_iter):
  gradf = 2*X.T@(X@w_hat[k]-y)
  w_hat[k+1] = w_hat[k] - tau*gradf
  f_hat[k+1] = la.norm(y-X@w_hat[k+1])**2


plt . plot (w.T, f, w_hat,f_hat,'-*',linewidth =2)
plt . title ('squared error loss for different candidate w')
plt . xlabel ('w')
plt . ylabel ('f(w)')
plt . ylim ([-.5,8.5])
plt . show ()
```

## Now with 2-d features

```python
n = 20
p = 2
X = np.matrix(np.random.rand(n, p))
w_true = np.matrix([[0.7],[0.5]])
y = X@w_true


N = 40
w1,w2 = np.meshgrid(np.linspace(0,1,N),np.linspace(0,1,N))
w = np.matrix([w1.flatten(),w2.flatten()])

f = np.square(la.norm(mat.repmat(y,1,N**2) - X@w,2,0))
f2 = np.reshape(f,[N,N])

plt.imshow(np.log(f2+1), extent=[0, 1, 0, 1], origin='lower')
plt.colorbar()



tau = .04
max_iter = 10
w_hat = np.matrix(np.zeros((p,max_iter+1)))
f_hat = np.matrix(np.zeros((max_iter+1,1)))
w_hat[:,0] = np.matrix([[0.05],[.45]])
f_hat[0] = la.norm(y-X@w_hat[:,0])**2

for k in range(max_iter):
  gradf = 2*X.T@(X@w_hat[:,k]-y)
  w_hat[:,k+1] = w_hat[:,k] - tau*gradf
  f_hat[k+1] = la.norm(y-X@w_hat[:,k+1])**2


plt.plot(w_hat[0,:].T,w_hat[1,:].T,'or-',linewidth =2)
plt . title ('points in feature space where gradients were evaluated')
plt . xlabel ('w_1')
plt . ylabel ('w_2')
plt . ylim ([0,1])
plt . xlim ([0,1])
plt . show ()
```
