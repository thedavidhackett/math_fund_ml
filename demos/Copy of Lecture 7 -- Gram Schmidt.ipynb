# Lecture 7, Gram-Schmidt

```python
import numpy as np
import scipy.linalg as sla
import numpy.linalg as la
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

plt.rcParams['figure.figsize'] = [8, 4]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
```

## create and plot data

```python
n = 2
p = 100
c = .01
Ustar = np.matrix([[1,c],[1,-c]])
Vstar = np.random.rand(n,p)-.5
noise = np.random.randn(n,p)/20

X = np.matrix(Ustar@Vstar + noise)

%matplotlib inline
fig = plt.figure()
ax = fig.add_subplot(111)

plt.scatter([X[0]],[X[1]])

i = np.random.randint(0,p)
j = np.random.randint(0,p)

U1 = X[:,i]
U1 = U1/la.norm(U1)
U2 = X[:,j]-U1@U1.T@X[:,j]
U2 = U2/la.norm(U2)

plt.plot((0,U1[0]),(0,U1[1]),'g',linewidth=3)
plt.plot((0,U2[0]),(0,U2[1]),'r',linewidth=3)
plt.scatter(X[0,i],X[1,i],c='k')
plt.scatter(X[0,j],X[1,j],c='r')


print(U1)
print(U2)
ax.set_aspect('equal', adjustable='box')


```
